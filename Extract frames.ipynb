```python
import cv2
import os
# type the address of the file which you store videos
video_path = 'video_2022year/'

## if you do not have package cv2 and os, you can use "conda install" to install all packages, details you can google it.
```

```python
videos = os.listdir(video_path)
print(videos)
```

```python
 for name in videos:
        video_name = name.split('.')[0]
        print(video_name)
```

```python
def get_image(video_path, gap):
    """
    @video_path is the input of the file including all the videos, remember to add "/" at the end of address.
    
    @gap, the type is int, meaning one frame will be extracted from "gap" frames.
    """
    videos = os.listdir(video_path)
    
    for name in videos:
        video_name = name.split('.')[0]
        folder_name = video_path + video_name
        os.makedirs(folder_name, exist_ok=True)
        print(folder_name)
        
        video = cv2.VideoCapture(video_path + name)
        index = 0
        rval = video.isOpened()
        
        while rval:
            index += 1
            rval, frame = video.read()
            
            if index % gap == 0:
                pic_dir = folder_name+'/'
                
                if rval:
                    anonym = str(video_name)+'_'+str(index)
                    cv2.imwrite(pic_dir+anonym+'.png', frame)
                    cv2.waitKey(1)
                    print('{} {} frame saved'.format(video_name, index))

                else:
                    break

        video.release()
        print('video is saved')
```

```python
get_image(video_path, 50)
```

```python
# If you have extract the frames, you will label them, then when you divide them into training dataset and testing dataset, 
# you need to extract corresponding label files, this is the use of this programm.

#just write your address of file store images 
path_pic = "D:/Mussel-reef/mydata/images/val_byBack"
path_lab = "D:/Mussel-reef/mydata/images/full/labels"
desPath = "D:/Mussel-reef/mydata/labels/val_byBack"

#store image names
pic_names = os.listdir(path_pic)
lab_names = os.listdir(path_lab)

#take picture names
for pic_name in pic_names:
    name = pic_name.split('.')[0]
    shutil.copy(path_lab+'/'+name+'.txt', desPath)
    
    print(pic_name+'is okay')
```

```python
#just write your address of file store labels
path  = 'D:\Mussel-reef\mydata\images\train_byBack'

desPath = path + 'images with labels'
desLabels = path + 'images with labels/labels'

#compare the name of pictures and label files, if same put them into same fie
files = os.listdir(path)
for file in files:
    path_1 = path+file+'/labels'
    os.makedirs(desPath, exist_ok=True)
    os.makedirs(desLabels, exist_ok=True)
    
    try:
        labels = os.listdir(path_1)
        print(path_1, '\n')
        
        for label in labels:
            name = label.split('.')[0]
            shutil.copy(path+file+'/'+name+'.png', desPath)
            shutil.copy(path_1+'/'+label, desLabels)
            print(path+file+name+'.png is finished')
            
    except:
        print('No labels in {}'.format(path_1), '\n')
        
        
```
